```python
import pandas as pd
df = pd.read_csv('proj_cleaned_encoded.csv')
df.head()
```

```python
from uszipcode import SearchEngine
search_zip = SearchEngine(simple_zipcode=True)
```

```python
def getInfo(zipcode):
    zipcode = int(zipcode)
    zipcode = str(zipcode)
    if len(zipcode) < 5:
        zipcode = '0'+ zipcode
    elif len(zipcode) < 4:
        zipcode = '00' + zipcode
    elif len(zipcode) < 3:
        zipcode = '000' + zipcode
    elif len(zipcode) < 2:
        zipcode = '0000' + zipcode
    print(zipcode)
    zipc = search_zip.by_zipcode(zipcode)
    z_dict = zipc.to_dict()
    #if z_dict.get('lat') is None or z_dict.get('lng') is None or z_dict.get('median_household_income') is None or z_dict.get('population') is None or z_dict.get('population_density') is None or z_dict.get('housing_units') is None or z_dict.get('occupied_housing_units') is None or z_dict.get('median_home_value') is None:
    #    return ['NaN', 'NaN'
    #else:
    return [z_dict.get('median_household_income'), z_dict.get('population'), z_dict.get('population_density'), z_dict.get('housing_units'), z_dict.get('occupied_housing_units'), z_dict.get('median_home_value')]
```

```python
newcolumns = df[0:10]['school_zip'].apply(getInfo)
```

```python
newcolumns
```

```python
newcolumns.apply(lambda x: x[0])
```

```python
newcolumns.apply(lambda x: x[2])
```

```python
newnewcolumns = df['school_zip'].apply(getInfo)
```

```python
['median_household_income', 'population', 'population_density', 'housing_units', 'occupied_housing_units', 'median_home_value']
```

```python
df['school_zip_median_household_income'] = newnewcolumns.apply(lambda x: x[0])
df['school_zip_population'] = newnewcolumns.apply(lambda x: x[1])
df['school_zip_pop_density'] = newnewcolumns.apply(lambda x: x[2])
df['school_zip_housing_units'] = newnewcolumns.apply(lambda x: x[3])
df['school_zip_occupied_housing_units'] = newnewcolumns.apply(lambda x: x[4])
df['school_zip_median_home_value'] = newnewcolumns.apply(lambda x: x[5])
```

```python
df.head()
```

```python
df.columns
```

```python
df.dropna(inplace=True)
```

```python
df['funding_status'].unique()
```

```python
df['funding_status'].replace(['failure', 'success'],[0,1],inplace=True)
df['funding_status'].unique()
```

```python
#X = df[['school_metro', 'school_charter',
#       'school_magnet', 'school_year_round', 'school_nlns', 'school_kipp',
#       'school_charter_ready_promise', 'teacher_prefix',
#       'teacher_teach_for_america', 'teacher_ny_teaching_fellow',
#       'primary_focus_subject', 'primary_focus_area', 'resource_type',
#       'poverty_level', 'grade_level',
#       'total_price_excluding_optional_support',
#       'total_price_including_optional_support', 'students_reached', 'eligible_double_your_impact_match',
#       'eligible_almost_home_match', 'timespan', 'month_posted',
#       'school_zip_median_household_income', 'school_zip_population',
#       'school_zip_pop_density', 'school_zip_housing_units',
#       'school_zip_occupied_housing_units', 'school_zip_median_home_value']]
X = df[['primary_focus_subject', 'primary_focus_area', 'resource_type',
       'poverty_level', 'grade_level','total_price_including_optional_support', 'students_reached', 'month_posted']]
y = df['funding_status']
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model
```

```python
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
X_train.head()
```

```python
predictions = model.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
print(X.columns)
model.coef_
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, predictions)
```

```python
print(classification_report(y_test, predictions))
```
