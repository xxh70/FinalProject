```python
def Average(lst):
    if len(lst) == 0:
        return 'NaN'
    else:
        return sum(lst) / len(lst) 

def getDistance(df):
    school_coords = [df['school_latitude'],df['school_longitude']]
    donor_zip = search_zip.by_city_and_state(df['donor_city'], df['donor_state'])
    lats = []
    lngs = []
    med_income = []
    pop = []
    pop_dens = []
    housing = []
    housing_occu = []
    med_home_val = []
    for c in donor_zip:
        z_dict = c.to_dict()
        if z_dict.get('lat') is None or z_dict.get('lng') is None or z_dict.get('median_household_income') is None or z_dict.get('population') is None or z_dict.get('population_density') is None or z_dict.get('housing_units') is None or z_dict.get('occupied_housing_units') is None or z_dict.get('median_home_value') is None:
            next
        else:
            lats.append(z_dict.get('lat'))
            lngs.append(z_dict.get('lng'))
            med_income.append(z_dict.get('median_household_income'))
            pop.append(z_dict.get('population'))
            pop_dens.append(z_dict.get('population_density'))
            housing.append(z_dict.get('housing_units'))
            housing_occu.append(z_dict.get('occupied_housing_units'))
            med_home_val.append(z_dict.get('median_home_value'))
    donor_coords = [Average(lats),Average(lngs)]
    if donor_coords[0] == 'NaN' or donor_coords[1] == 'NaN':
        return ['NaN', Average(med_income), Average(pop), Average(pop_dens), Average(housing), Average(housing_occu), Average(med_home_val)]
    else:
        return [gp.distance(school_coords, donor_coords).mi,Average(med_income),Average(pop), Average(pop_dens), Average(housing), Average(housing_occu), Average(med_home_val)]
```

```python
newcoltest = df[0:10].apply(getDistance, axis=1)
```
