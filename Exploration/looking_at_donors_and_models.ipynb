```python
def Average(lst):
    if len(lst) == 0:
        return 'NaN'
    else:
        return sum(lst) / len(lst) 

def getCoordsAndMedIncome(city,state):
    z = search_zip.by_city_and_state(city, state)
    lats = []
    lngs = []
    med_income = []
    for c in z:
        z_dict = c.to_dict()
        if z_dict.get('lat') is None or z_dict.get('lng') is None or z_dict.get('median_household_income') is None:
            next
        else:
            lats.append(z_dict.get('lat'))
            lngs.append(z_dict.get('lng'))
            med_income.append(z_dict.get('median_household_income'))
    return([Average(lats),Average(lngs),Average(med_income)])
```
